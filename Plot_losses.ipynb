```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
f = open('losses.pkl', 'rb')
losses_dict = pickle.load(f)
```

```python
train_losses = losses_dict['train_loss']
val_losses = losses_dict['val_losses'][1:]
```

```python
assert(len(train_losses)==len(val_losses), "Length of losses list should be equal for train and validation")
```

```python
epochs = list(range(1, len(train_losses) + 1))
```

```python
# https://stackabuse.com/change-font-size-in-matplotlib/
fig, ax = plt.subplots(figsize=(6, 4))

ax.plot(epochs, train_losses, '-D', label='Training losses')
ax.plot(epochs, val_losses, '-D', label='Validation losses')
ax.set_xlabel('epochs', fontsize=15)
ax.set_ylabel('losses', fontsize=15)
leg = ax.legend(prop={"size":12})
ax.grid()

plt.show()
```
